```python
# Import dependencies
import pandas as pd
```

```python
# Import csv into DataFrame
housing_data_df = pd.read_csv('austinHousingData.csv')
```

```python
# Make a working copy of DataFrame
df = housing_data_df.copy()
df
```

```python
# Display all columns of the DataFrame 
pd.options.display.max_columns = None
```

```python
display(df)
```

```python
# Check data types
df.info()
```

```python
# Remove unnecessary columns
clean_df = df[['zpid','city','streetAddress','zipcode',
            'latitude','longitude','homeType','yearBuilt',
            'latestPrice','numPriceChanges','numOfPrimarySchools','numOfElementarySchools',
             'numOfMiddleSchools','numOfHighSchools', 'avgSchoolDistance']]
clean_df
```

```python
# Drop any null values
clean_df.dropna()
```

```python
# Drop any duplicate values
clean_df.drop_duplicates()
```

```python
# Suppress Scientific Notation
pd.options.display.float_format = '{:.2f}'.format

# Group average home price by zipcode
avg_price = clean_df.groupby(['zipcode']).mean()['latestPrice']
avg_price
```

```python
# Create new DataFrame with zipcodes and average home price
avg_df = pd.DataFrame({
    'Average Price' : avg_price
})

# Sort the new DataFrame
avg_df.sort_values(by='Average Price', ascending=True)
```

```python
# View average price change by zipcode
avg_price_change = clean_df.groupby(['zipcode']).mean()['numPriceChanges']
avg_price_change
```
